### Elementwise addition

```python
np8x4_1 = np.add(np8x_, np8x_)
np8x4_1
```

### Elementwise Multiply

```python
np8x_ = np.arange(6).reshape(3,-1)
np8x_
```

```python
np8x_mul = np8x_ * np8x_
np8x_mul
```

### Matrix Multiply

```python
np2x3 = np8x_mul.reshape(2,3)
np2x3
```

```python
np8x_mul
```

```python
np8x_matmul = np8x_mul @ np2x3
np8x_matmul
```

```python
np8x_matmul =  np2x3 @ np8x_mul
np8x_matmul
```
